```python
import pandas as pd

# Load the dataset
url = 'https://raw.githubusercontent.com/KeithGalli/pandas/master/pokemon_data.csv'
df = pd.read_csv(url)

# Total number of rows and columns
total_rows, total_columns = df.shape
print(f"Total rows: {total_rows}, Total columns: {total_columns}")

# Descriptive statistics
print("\nDescriptive Statistics:")
describe_output = df.describe()
print(describe_output)

# Count of non-null values in each column
print("\nCount of non-null values:")
print(df.count())

# Missing values in numeric columns
print("\nMissing values in numeric columns:")
print(df.isnull().sum())
```

Count:

Definition: The number of non-null (non-missing) values in the column.
Purpose: Helps to determine how many values are present for the given column, excluding any missing values.
Mean:

Definition: The arithmetic average of the values in the column.
Calculation: Sum of all values divided by the count of values.
Purpose: Provides a measure of the central tendency of the data. It indicates the average value in the column.
Standard Deviation (std):

Definition: A measure of the amount of variation or dispersion of the values from the mean.
Calculation: The square root of the average of the squared differences between each value and the mean.
Purpose: Shows how spread out the values are around the mean. A higher standard deviation indicates more variability.
Minimum (min):

Definition: The smallest value in the column.
Purpose: Provides the lower boundary of the data range. Useful for understanding the minimum value recorded in the dataset.
25th Percentile (25%):

Definition: The value below which 25% of the data falls.
Purpose: Helps to understand the distribution of the data. It’s a measure of the lower quartile and gives insight into the lower range of values.
50th Percentile (50%):

Definition: The median value of the column, where 50% of the data falls below this value and 50% falls above.
Purpose: Provides a measure of the central value of the data, splitting it into two equal halves.
75th Percentile (75%):

Definition: The value below which 75% of the data falls.
Purpose: Helps to understand the distribution of the data. It’s a measure of the upper quartile and gives insight into the upper range of values.
Maximum (max):

Definition: The largest value in the column.
Purpose: Provides the upper boundary of the data range. Useful for understanding the maximum value recorded in the dataset.
