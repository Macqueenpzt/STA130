1.In mathematical form, the model is:Y=Œ≤0+Œ≤1X+œµ

Y: Outcome variable, whose values we want to predict based on X

X: Predictor variable, which influences ùëå

Œ≤0: Intercept coefficient, the value of Y when X = 0

Œ≤1: Slope coefficient, which represents the change in Y when X changes. It captures the strength and direction of the relationship.

ùúñ: This error term is assumed to follow a normal distribution, ùúñ‚àºùëÅ(0,œÉ2) where œÉ is the standard deviation.

Y‚àºN(Œ≤0+Œ≤1X,œÉ2), where œÉ2 is the square of standard deviation, the outcome variable Y at each value of X is thus drawn from a normal distribution with mean Œ≤0+Œ≤1X and standard deviation œÉ.

````python
# the link of using ChatGPT: https://chatgpt.com/share/6717b395-55c8-8008-b611-a0734268b276
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Parameters for the regression model
beta_0 = 2.0  # intercept
beta_1 = 0.5  # slope
sigma = 1.0   # standard deviation of the error term

# Generate random values for X (the predictor)
np.random.seed(42)  # for reproducibility
X = np.random.uniform(0, 10, 100)

# Generate normally distributed errors (epsilon)
epsilon = np.random.normal(0, sigma, size=X.shape)

# Calculate the outcome variable Y
Y = beta_0 + beta_1 * X + epsilon

# Visualize the data points (Y) and the true regression line
plt.scatter(X, Y, label='Sampled Data', color='blue', alpha=0.6)
plt.plot(X, beta_0 + beta_1 * X, color='red', label='True Regression Line')
plt.title('Simple Linear Regression: Data with True Line')
plt.xlabel('Predictor Variable (X)')
plt.ylabel('Outcome Variable (Y)')
plt.legend()
plt.show()

# Perform a simple linear regression using scipy.stats
slope, intercept, r_value, p_value, std_err = stats.linregress(X, Y)

# Print out the estimated slope and intercept from the regression
print(f"Estimated slope: {slope}")
print(f"Estimated intercept: {intercept}")
````

````python
# 2.
# the link of using ChatGPT: https://chatgpt.com/share/6717b395-55c8-8008-b611-a0734268b276
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

# Step 1: Simulate the dataset
np.random.seed(42)  # for reproducibility
X = np.random.uniform(0, 10, 100)  # 100 random values for X
beta_0 = 2.0  # true intercept
beta_1 = 0.5  # true slope
sigma = 1.0   # standard deviation of the error
epsilon = np.random.normal(0, sigma, 100)  # error term
Y = beta_0 + beta_1 * X + epsilon  # outcome variable

# Step 2: Create a pandas DataFrame for easier handling
df = pd.DataFrame({'X': X, 'Y': Y})

# Step 3: Fit the Simple Linear Regression model using statsmodels
# Formula notation 'Y ~ X' means we're modeling Y as a function of X
model = smf.ols(formula='Y ~ X', data=df)
results = model.fit()

# Step 4: Print out the summary of the fitted regression model
print(results.summary())

# Step 5: Visualize the fitted model
# Generate predictions from the fitted model
df['predicted_Y'] = results.fittedvalues

# Plot the original data points
plt.scatter(df['X'], df['Y'], label='Sampled Data', color='blue', alpha=0.6)

# Plot the fitted regression line
plt.plot(df['X'], df['predicted_Y'], color='red', label='Fitted Line')

# Customize the plot
plt.title('Simple Linear Regression: Fitted Line vs Sampled Data')
plt.xlabel('Predictor Variable (X)')
plt.ylabel('Outcome Variable (Y)')
plt.legend()

# Show the plot
plt.show()
````

````python
# the link of using ChatGPT: https://chatgpt.com/share/6717b395-55c8-8008-b611-a0734268b276
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf  # for specifying and fitting the regression model
import plotly.express as px  # for creating interactive plots

# Step 1: Simulate the dataset
np.random.seed(42)  # for reproducibility
X = np.random.uniform(0, 10, 100)  # 100 random values for X
beta_0 = 2.0  # true intercept
beta_1 = 0.5  # true slope
sigma = 1.0   # standard deviation of the error
epsilon = np.random.normal(0, sigma, 100)  # error term
Y = beta_0 + beta_1 * X + epsilon  # outcome variable (y)

# Step 2: Create a pandas DataFrame with column names 'x' and 'y'
df = pd.DataFrame({'x': X, 'y': Y})

# Step 3: Specify and fit the Simple Linear Regression model using statsmodels
model_data_specification = smf.ols("y~x", data=df)  # Specify the model 'y ~ x'
fitted_model = model_data_specification.fit()  # Fit the model to the data

# Step 4: Visualize the data and fitted trendline using plotly
df['Data'] = 'Data'  # Add this column to create a label for the legend

# Create a scatter plot of the data points, with an OLS trendline
fig = px.scatter(df, x='x', y='y', color='Data', trendline='ols', title='y vs. x')

# This step explicitly adds the trendline (the fitted regression line)
fig.add_scatter(x=df['x'], y=fitted_model.fittedvalues,
                line=dict(color='blue'), name="OLS Trendline")

# Step 5: Calculate the true regression line and add it to the plot
true_y = beta_0 + beta_1 * df['x']  # Calculate y values based on true parameters
fig.add_scatter(x=df['x'], y=true_y,
                line=dict(color='red', dash='dash'), name='True Regression Line')

# Show the plot (for GitHub/MarkUs submissions, use renderer="png")
fig.show(renderer="png")  # Use this if you're submitting to GitHub or MarkUs
````

The OLS method is used to find the best-fitting line in a linear regression model. OLS chooses the line that minimizes the sum of squared residuals because squaring the residuals ensures that all deviations (both positive and negative) are treated equally, and larger deviations are penalized more. This method yields the fitted line that best represents the data, based on the observed data and considering the variability present in the dataset.The fitted line is chosen because it minimizes the overall discrepancy between the observed data points and the line, using the sum of squared residuals as a metric. This approach ensures that the line is the best possible fit for the given data, balancing out small and large errors effectively.

And the reasons why it requires square is by squaring the residuals, OLS ensures that all residuals (whether positive or negative) contribute equally to the sum. Otherwise, positive and negative residuals could cancel each other out.

````python
import scipy.stats as stats
import pandas as pd
import statsmodels.formula.api as smf
import plotly.express as px
import numpy as np

# Define parameters for the data generation
n, x_min, x_range, beta0, beta1, sigma = 20, 5, 5, 2, 3, 5

# Generate random x values uniformly within the specified range
x = stats.uniform(x_min, x_range).rvs(size=n)

# Generate random errors (normally distributed with mean 0 and std deviation sigma)
errors = stats.norm(loc=0, scale=sigma).rvs(size=n)

# Create the y values using the true theoretical relationship: y = beta0 + beta1 * x + error
y = beta0 + beta1 * x + errors

# Combine x and y into a DataFrame for analysis
df = pd.DataFrame({'x': x, 'y': y})

# Specify and fit the OLS regression model to the data
model_data_specification = smf.ols("y~x", data=df)  # OLS: y as a function of x
fitted_model = model_data_specification.fit()  # Fit the model to find the best-fitting line

# Add a column for data points (just for legend purposes in the plot)
df['Data'] = 'Data'

# Create a scatter plot of x and y and add the fitted OLS line (blue line)
fig = px.scatter(df, x='x', y='y', color='Data', trendline='ols', title='y vs. x')

# Manually add the OLS trendline based on the fitted values from the model
fig.add_scatter(x=df['x'], y=fitted_model.fittedvalues, line=dict(color='blue'), name="trendline='ols'")

# Calculate the theoretical line y = beta0 + beta1 * x (true line without noise)
x_range = np.array([df['x'].min(), df['x'].max()])
y_line = beta0 + beta1 * x_range

# Add the true theoretical line (orange dashed line) for comparison
fig.add_scatter(x=x_range, y=y_line, mode='lines', name=str(beta0)+' + '+str(beta1)+' * x', 
                line=dict(dash='dot', color='orange'))

# Add vertical dashed red lines to represent the residuals (differences between observed and fitted values)
for i in range(len(df)):
    fig.add_scatter(x=[df['x'][i], df['x'][i]],  # x remains the same
                    y=[fitted_model.fittedvalues[i], df['y'][i]],  # y goes from fitted value to observed value
                    mode='lines',
                    line=dict(color='red', dash='dash'),
                    showlegend=False)

# Add a horizontal dashed black line at the mean of the y-values (y-bar) for reference
fig.add_scatter(x=x_range, y=[df['y'].mean()]*2, mode='lines',
                line=dict(color='black', dash='dot'), name='y-bar')

# Show the plot
fig.show(renderer = "png")  # Use fig.show(renderer="png") if submitting on GitHub or MarkUs
````

````python
#B.
# Load the dataset
old_faithful = sns.load_dataset('geyser')

# Filter for long wait times (adjust this threshold as needed)
long_wait = old_faithful.waiting >= 60  # Example threshold for long wait times
data_long_wait = old_faithful[long_wait]

# Set parameters for the simulation
beta_0 = 1.65
beta_1 = 0
sigma = 0.37
n = len(data_long_wait)  # Use the actual number of entries for long wait times

# Generate x_values and simulate y values
x_values = data_long_wait['waiting'].values
y_simulated = beta_0 + beta_1 * x_values + np.random.normal(0, sigma, size=len(x_values))  # Match size with x_values

# Fit the model to the simulated data
simulated_slopes = []
for _ in range(1000):  # Adjust number of bootstraps as needed
    sample_indices = np.random.choice(range(len(x_values)), size=n, replace=True)
    y_sampled = y_simulated[sample_indices]
    x_sampled = x_values[sample_indices]
    model = smf.ols('y_sampled ~ x_sampled', data=pd.DataFrame({'y_sampled': y_sampled, 'x_sampled': x_sampled})).fit()
    simulated_slopes.append(model.params['x_sampled'])

# Convert to DataFrame for visualization
simulated_slopes_df = pd.DataFrame(simulated_slopes, columns=['simulated_slope'])

# Visualize the simulated sampling distribution
fig2 = px.histogram(simulated_slopes_df, x='simulated_slope', nbins=30, title='Simulated Sampling Distribution of Slope Coefficients (Null Hypothesis)')
fig2.show(renderer="png")
````

Frist, repeatedly shuffling the group labels and calculating the mean difference in duration to see how extreme our observed mean difference is relative to the distribution of differences under the null hypothesis. Second, resampling each group separately with replacement and computing the difference in sample means to create a 95% confidence interval for the difference in population means.

````python
# the link of using ChatGPT: https://chatgpt.com/share/671fb347-9eb0-8008-ad28-58ba34daa136
import numpy as np
import pandas as pd
from scipy import stats
import statsmodels.formula.api as smf
import seaborn as sns

# Load dataset
old_faithful = sns.load_dataset('geyser')
observed_mean_diff = old_faithful.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]
observed_mean_diff
np.random.seed(42)
num_permutations = 1000
permuted_diffs = []

for _ in range(num_permutations):
    # Shuffle the 'kind' labels
    shuffled_labels = old_faithful['kind'].sample(frac=1, replace=False).values
    # Assign shuffled labels back to a new column
    old_faithful['kind_shuffled'] = shuffled_labels
    # Calculate the difference in means with shuffled labels
    mean_diff = old_faithful.groupby('kind_shuffled')['duration'].mean().iloc[::-1].diff().values[1]
    permuted_diffs.append(mean_diff)
num_bootstrap_samples = 1000
bootstrapped_diffs = []

for _ in range(num_bootstrap_samples):
    # Resample each group separately with replacement
    boot_sample = old_faithful.groupby('kind').apply(lambda x: x.sample(n=len(x), replace=True)).reset_index(drop=True)
    # Calculate the difference in means for the bootstrap sample
    boot_mean_diff = boot_sample.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]
    bootstrapped_diffs.append(boot_mean_diff)

# Calculate 95% confidence interval from bootstrap differences
ci_95 = np.quantile(bootstrapped_diffs, [0.025, 0.975])
ci_95
````

In summary, the indicator variable model provides a structured way to assess differences, but relies on model assumptions. The permutation test is useful for hypothesis testing without assumptions, while bootstrap CI gives an interval estimate for the mean difference. Each method offers insights, with the permutation test and bootstrap CI being more robust to deviations from normality.

> The first three assumptions associated with the **Simple Linear regression** model are that<br>
> 
> 1. the $\epsilon_i$ **errors** (sometimes referred to as the **noise**) are **normally distributed**

> 3. the linear form is [at least reasonably approximately] "true" (in the sense that the above two remain [at least reasonably approximately] "true") so that then behavior of the $Y_i$ **outcomes** are represented/determined on average by the **linear equation**)<br>
> 
>    and there are additional assumptions; but, a deeper reflection on these is "beyond the scope" of STA130; nonetheless, they are that<br><br>
> 4. the $x_i$ **predictor variable** is **measured without error**
> 5. and the $\epsilon_i$ **errors** are **statistically independent** (so their values do not depend on each other)
> 6. and the $\epsilon_i$ **errors** are **unbiased** relative to the **expected value** of **outcome** $E[Y_i|x_i]=\beta_0 + \beta_1x_i$ (which is equivalently stated by saying that the mean of the **error distribution** is $0$, or again equivalently, that the **expected value** of the **errors** $E[\epsilon_i] = 0$)
