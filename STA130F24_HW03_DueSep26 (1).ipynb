Load the Penguins dataset from the provided URL.

Process the flipper_length_mm data for each species (mean, median, range, interquartile range, and two standard deviations).

Generate Plotly histograms for the flipper_length_mm of each species.

Add vertical lines (add_vline) to mark the mean and median.

Add vertical shaded regions (add_vrect) to highlight ranges: the overall range, interquartile range, and two standard deviations around the mean.

````python
import pandas as pd
import plotly.graph_objects as go

# Load the dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv"
penguins = pd.read_csv(url)

# Filter out rows with missing values in 'flipper_length_mm'
penguins = penguins.dropna(subset=['flipper_length_mm'])

# Prepare the figure
fig = go.Figure()

# Define a function to add histograms and lines/rectangles
def add_species_histogram(species_data, species_name):
    # Calculate statistics
    mean = species_data['flipper_length_mm'].mean()
    median = species_data['flipper_length_mm'].median()
    std_dev = species_data['flipper_length_mm'].std()
    min_val = species_data['flipper_length_mm'].min()
    max_val = species_data['flipper_length_mm'].max()
    q1 = species_data['flipper_length_mm'].quantile(0.25)
    q3 = species_data['flipper_length_mm'].quantile(0.75)
    
    # Histogram
    fig.add_trace(go.Histogram(
        x=species_data['flipper_length_mm'],
        name=species_name,
        opacity=0.75,
        histnorm='probability density'))

    # Add lines for mean and median
    fig.add_shape(type='line', x0=mean, y0=0, x1=mean, y1=0.1, line=dict(color='red', width=2), name='Mean')
    fig.add_shape(type='line', x0=median, y0=0, x1=median, y1=0.1, line=dict(color='green', width=2), name='Median')
    
    # Add rectangles for range and IQR
    fig.add_shape(type='rect', x0=min_val, y0=0, x1=max_val, y1=0.1, line=dict(color='blue', width=0), fillcolor='blue', opacity=0.2, name='Range')
    fig.add_shape(type='rect', x0=q1, y0=0, x1=q3, y1=0.1, line=dict(color='orange', width=0), fillcolor='orange', opacity=0.3, name='IQR')

    # Add lines for ±2 standard deviations
    fig.add_shape(type='line', x0=mean - 2*std_dev, y0=0, x1=mean - 2*std_dev, y1=0.1, line=dict(color='purple', width=2, dash='dash'), name='-2 SD')
    fig.add_shape(type='line', x0=mean + 2*std_dev, y0=0, x1=mean + 2*std_dev, y1=0.1, line=dict(color='purple', width=2, dash='dash'), name='+2 SD')
    
    # Loop through each species
for species in penguins['species'].unique():
    species_data = penguins[penguins['species'] == species]
    add_species_histogram(species_data, species)

# Update layout
fig.update_layout(
    title='Flipper Length Distribution by Species',
    xaxis_title='Flipper Length (mm)',
    yaxis_title='Density',
    barmode='overlay',
    showlegend=False)

# Show the figure
fig.show(renderer="png")
````

To approach this task, we'll use Seaborn to generate Kernel Density Estimation (KDE) plots for each species' flipper_length_mm, and we'll plot these KDE curves onto a 3-row layout, each corresponding to a species from the penguins dataset. We'll also mark:

The mean and median locations with vertical lines.
The scale indicators such as:
The overall range (min and max),
The interquartile range (IQR),
The range defined by two standard deviations from the mean.
Steps:
Load and process the Penguins dataset.
Calculate statistics for flipper_length_mm for each species.
Use Seaborn's KDE plots to visualize the distributions.
Mark the calculated statistics (mean, median, ranges) onto the KDE plots.
Organize the plots in a 3-row layout (one per species).

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the penguins dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv"
df = pd.read_csv(url)

# Group data by species and calculate necessary statistics for flipper_length_mm
stats = df.groupby('species')['flipper_length_mm'].agg(
    mean='mean',
    median='median',
    min_val='min',
    max_val='max',
    std='std',
    q25=lambda x: x.quantile(0.25),
    q75=lambda x: x.quantile(0.75)
).reset_index()

# Add range based on two standard deviations
stats['mean_minus_2std'] = stats['mean'] - 2 * stats['std']
stats['mean_plus_2std'] = stats['mean'] + 2 * stats['std']

# Initialize the figure with 3 subplots (one for each species)
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)
species_list = df['species'].unique()

# Iterate through each species and plot KDE
for i, species in enumerate(species_list):
    ax = axes[i]
    species_data = df[df['species'] == species]
    species_stats = stats[stats['species'] == species].iloc[0]
    
    # KDE plot for flipper_length_mm
    sns.kdeplot(species_data['flipper_length_mm'], fill=True, ax=ax, label=f'{species}', color='blue')
    
    # Add vertical lines for mean and median
    ax.axvline(species_stats['mean'], color='blue', linestyle='--', label='Mean')
    ax.axvline(species_stats['median'], color='green', linestyle='--', label='Median')
    
    # Add shaded regions for range, IQR, and two standard deviations
    ax.axvspan(species_stats['min_val'], species_stats['max_val'], color='lightblue', alpha=0.2, label='Range')
    ax.axvspan(species_stats['q25'], species_stats['q75'], color='lightgreen', alpha=0.4, label='IQR')
    ax.axvspan(species_stats['mean_minus_2std'], species_stats['mean_plus_2std'], color='orange', alpha=0.3, label='2 Std. Dev.')

    # Set titles and labels
    ax.set_title(f"KDE for {species}")
    ax.set_xlabel("Flipper Length (mm)")
    if i == 0:
        ax.set_ylabel("Density")

    # Add legend
    ax.legend(loc="upper right")

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
````

Box plots are graphical charts that can represent the spread and variance of numerical data distribution around its mean value. Its effect is qimilar to bar plots and histograms, but it provides a higher level of statistics data. The "box" of a box plot is constructed out of the first (Q1) and third quartiles (Q3) separated by the median (which is the second quartile Q2); so, a box plot shows the the 25th, 50th, and 75th percentile with the "box" containing the middle 50% of the data. Box plots are efficient for comparing distribution of different group. Besides, it is easily to view the skewness of data through box plots. But the disadvantage is that we cannot view the multimodality and sample size (data count) by using box plots.

Histograms use bars of different heights, which represent how many data points fall into this range. The bins(width) of histogrms represent each interval of the data. It might be the best method for visualizing the overall distribution of a single variable. Histograms' advantages are that it shows both multimodality and sample size. However, the effects of histograms largely depends on the size of bins, if the bins size is too big, there are too many data fall in each interval, resulting in that we cannot easily see details about how the data is distributed. If the size is too small, histograms have too many bars and it might be hard to view the overall distribution.

A kernel density estimators are non-parametric ways to estimate the probability density function of a random variable. Unlike histograms, KDE smoothens the distribution curve, providing a continuous estimate of the density function.

Answer from ChatGPT: The mean and median are both measures of central tendency, but they capture different aspects of a dataset's distribution. Here's a breakdown of their general relationship:

1. Definition:

Mean: The arithmetic average of all values. It is calculated by summing all the values and dividing by the number of values.

Median: The middle value of a dataset when it is ordered. If the dataset has an odd number of observations, the median is the middle number. If it's even, it's the average of the two middle numbers.

2. Symmetrical Distributions:

In a symmetrical distribution (e.g., a normal distribution):The mean and median are the same because the data is evenly spread around a central point.Mean=Median

3. Skewed Distributions:

When a dataset is skewed, the relationship between the mean and median changes:Right-skewed (positively skewed): The tail on the right side is longer. In this case, the mean is typically greater than the median because the higher values "pull" the mean to the right. Mean > Median

Left-skewed (negatively skewed): The tail on the left side is longer. Here, the mean is usually less than the median because the lower values "pull" the mean to the left. Mean < Median


My understanding and explanation: mean is the average value and it is affeted by extreme values. Median is the middle value of dataset and it is not affected by extreme values. In a symmetrical distribution, mean is equal to median. In a skewed distribution, mean value is pulled in the direction of skewness, and median is closer to the middle position. Therefore, in a right skewness distribution, mean is greater than median; in a left skewness distribution, mean is less than median. In summary, in skewed distributions: The mean is sensitive to extreme values and tends to be pulled towards the tail with more extreme values. The median is less sensitive to extreme values and tends to be more robust, providing a better representation of the central tendency of the data.

````python
# Import necessary libraries
from scipy import stats
import pandas as pd
import numpy as np
import plotly.express as px

# Generate a right-skewed gamma distribution sample
sample1 = stats.gamma(a=2, scale=2).rvs(size=1000)

# Plot the histogram for the right-skewed distribution
fig1 = px.histogram(pd.DataFrame({'data': sample1}), x="data", title="Right-Skewed Gamma Distribution")
fig1.show(renderer="png")  # Use this renderer for GitHub and MarkUs submissions
# Calculate the mean and median of the right-skewed sample
mean1 = sample1.mean()
median1 = np.quantile(sample1, 0.5)

# Display the results
mean1, median1
# Generate a left-skewed gamma distribution sample (negated values)
sample2 = -stats.gamma(a=2, scale=2).rvs(size=1000)

# Plot the histogram for the left-skewed distribution
fig2 = px.histogram(pd.DataFrame({'data': sample2}), x="data", title="Left-Skewed Gamma Distribution")
fig2.show(renderer="png")  # Use this renderer for GitHub and MarkUs submissions
# Calculate the mean and median of the left-skewed sample
mean2 = sample2.mean()
median2 = np.quantile(sample2, 0.5)

# Display the results
mean2, median2
# Interpretation of the results

print(f"Right-Skewed Distribution: Mean = {mean1}, Median = {median1}")
print(f"In a right-skewed distribution, the mean is typically greater than the median.")

print(f"Left-Skewed Distribution: Mean = {mean2}, Median = {median2}")
print(f"In a left-skewed distribution, the mean is typically less than the median.")
````
