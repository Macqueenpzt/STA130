1(a) A Classification Decision Tree is a type of supervised learning algorithm used to address classification problems, where the goal is to predict a categorical or discrete target variable based on a set of input features. It predict the outcomes(most are categorical variables) based on sequential application of rules relative specific predictor variables. Most classification decision tree is called binary decision trees because it has two decision rules, the "left" one and the "right" one.

Real-word applications: 1. Medical Diagnosis

Problem: Predicting whether a patient has a specific disease based on symptoms, test results, or medical history.

Example: Classifying whether a tumor is malignant or benign using features like size, texture, and cell shape.

2.Species Classification in Ecology

Problem: Identifying species based on observable characteristics.

Example: Classifying plant species based on features like leaf size, color, and habitat.

1(b) In most cases Classification Decision Tree makes predictions about categorical variable. The tree splits the data into subsets based on feature values using if-then rules(decision rules). At each split, it chooses the feature and threshold that best separate the data into homogeneous groups until yielding outputs, which are often class label or class probabilities. Decision tree is used in classification problems field.

Multiple linear regression is used to make predictions about continuous variable and we first make assumptions that the outcome Yi has a linear correlation with predictor variables x1, x2...xn. Then use OLS obtain a formula for the fitted model Yi = β0 + β1x1 + β2x2 + ... + βnxn. And in most cases linear regression yields numeric data as outcomes.

1. **Accuracy** measures the proportion of true results (both true positives and true negatives) in the population.

$$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$$
    
2. **Sensitivity** measures the proportion of actual positives that are correctly identified.

$$\text{Sensitivity} = \frac{TP}{TP + FN}$$

3. **Specificity** measures the proportion of actual negatives that are correctly identified.

$$\text{Specificity} = \frac{TN}{TN + FP}$$

4. **Precision** measures the proportion of positive identifications that were actually correct.

$$\text{Precision} = \frac{TP}{TP + FP}$$

1.Accuracy. Weather Prediction Systems.

In predicting weather conditions (e.g., "Rain" or "No Rain"), both correct positive (rain forecast) and correct negative (no rain forecast) predictions are equally important for reliability. Accuracy is applied in the cases such that the True positive and True negative are equally significant, namely we focus on the true outcomes(both positive and negative). Accuracy works well when the cost of false positives (predicting rain when it doesn’t) and false negatives (missing a rain forecast) are similar.

2.Sensitivity. Cancer Detection.

In medical tests, identifying as many actual cases of disease (positives) as possible is critical to ensuring early diagnosis and treatment. High sensitivity reduces the risk of missing patients who have the disease (false negatives), which can have severe consequences.

3.Specificity. Fraud Detection in Financial Transactions.

For fraud detection, correctly identifying non-fraudulent transactions (negatives) minimizes disruptions for legitimate users. High specificity ensures fewer false positives (flagging legitimate transactions as fraudulent), maintaining customer trust and minimizing unnecessary interventions.

4.Precision. Spam Email Filtering.
In spam detection, precision ensures that flagged emails are truly spam to avoid mistakenly classifying important emails as spam. High precision reduces false positives, ensuring legitimate emails aren’t blocked unnecessarily, which could result in missed critical communications.

Summary: Accuracy is applied when TP and TN are equally important. Sensitivity is applied when we want to minimize FN. Specificity is applied when we want to minimize FP, avoiding false positives among actual negative cases and ensuring that negatives are correctly identified. Precision is applied when we want to minimize FP, avoiding false positives within predicted positives and ensuring the positive predictions are correct.

````python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score
import seaborn as sns
import matplotlib.pyplot as plt

# Load and preprocess the data
url = "https://raw.githubusercontent.com/pointOfive/STA130_F23/main/Data/amazonbooks.csv"
ab = pd.read_csv(url, encoding="ISO-8859-1")
ab_reduced = ab.drop(columns=["Weight_oz", "Width", "Height"])  # Drop unnecessary columns
ab_reduced_noNaN = ab_reduced.dropna()  # Remove rows with missing data

# Perform an 80/20 train-test split
train_data, test_data = train_test_split(ab_reduced_noNaN, test_size=0.2, random_state=42)

# Extract features and target variable for clf
X = ab_reduced_noNaN[['List Price']]  # Features for clf
y = pd.get_dummies(ab_reduced_noNaN['Hard_or_Paper'])['H']  # Target variable

# Train clf (model 1)
clf = DecisionTreeClassifier(max_depth=2, random_state=42)
clf.fit(X, y)

# Train clf2 (model 2) using multiple features
X2 = ab_reduced_noNaN[['NumPages', 'Thick', 'List Price']]  # Features for clf2
clf2 = DecisionTreeClassifier(max_depth=4, random_state=42)
clf2.fit(X2, y)

# Evaluate models on test data
# Features and target for test data
X_test = test_data[['List Price']]  # Features for clf
y_test = pd.get_dummies(test_data['Hard_or_Paper'])['H']  # Target variable

X_test2 = test_data[['NumPages', 'Thick', 'List Price']]  # Features for clf2

# Predictions
y_pred_clf = clf.predict(X_test)  # Predictions for clf
y_pred_clf2 = clf2.predict(X_test2)  # Predictions for clf2

# Confusion matrices
cm_clf = confusion_matrix(y_test, y_pred_clf)
cm_clf2 = confusion_matrix(y_test, y_pred_clf2)

# Plot confusion matrices
def plot_confusion_matrix(cm, title, labels):
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title(title)
    plt.show()

# Plot confusion matrix for clf
plot_confusion_matrix(cm_clf, "Confusion Matrix for clf", ["Paperback", "Hardcover"])

# Plot confusion matrix for clf2
plot_confusion_matrix(cm_clf2, "Confusion Matrix for clf2", ["Paperback", "Hardcover"])

# Metrics for clf
tn_clf, fp_clf, fn_clf, tp_clf = cm_clf.ravel()
sensitivity_clf = recall_score(y_test, y_pred_clf)  # True positive rate
specificity_clf = tn_clf / (tn_clf + fp_clf)  # True negative rate
accuracy_clf = accuracy_score(y_test, y_pred_clf)

# Metrics for clf2
tn_clf2, fp_clf2, fn_clf2, tp_clf2 = cm_clf2.ravel()
sensitivity_clf2 = recall_score(y_test, y_pred_clf2)  # True positive rate
specificity_clf2 = tn_clf2 / (tn_clf2 + fp_clf2)  # True negative rate
accuracy_clf2 = accuracy_score(y_test, y_pred_clf2)

# Report results
print("Confusion Matrix for clf:")
print(cm_clf)
print(f"Sensitivity: {sensitivity_clf:.3f}, Specificity: {specificity_clf:.3f}, Accuracy: {accuracy_clf:.3f}\n")

print("Confusion Matrix for clf2:")
print(cm_clf2)
print(f"Sensitivity: {sensitivity_clf2:.3f}, Specificity: {specificity_clf2:.3f}, Accuracy: {accuracy_clf2:.3f}")
````

The result for Specificity is 0.909, indicating that 90.9% proportion of actual negatives(Paperback) are correctly identified, Specificity = TN / (TN + FP) = 40/(40+4) = 40/44 = 0.909.

Therefore, the Sensitivity, Specifcity and Accuracy for the model clf are all close to 1, then the model makes largely accurate predictions and performs well.

<details class="details-example"><summary style="color:blue"><u>Further Guidance</u></summary>
  
> The way a **classification decision tree** is fit is that at each step in the construction process of adding a new **decision node splitting rule** to the current **tree structure**, all possible **decision rules** for all possible **predictor variables** are considered, and the combination that improves the **prediction** the most (as measured by the criterion of either "Gini impurity" or "Shannon entropy") and in accordance with the rules of the decision tree (such as the `max_depth` argument) is added to the **classification decision tree**.  Thus overall "criterion" noted above improves with each new **decision node splitting rule**, so the improvement can thus be tracked and the improvement contributions attributed to the **feature** upon which the **decision node splitting rule** is based.  This means the relative contribution of each **predictor variable** to the overall explanatory power of the model can be calculated, and this is what the `.feature_importances_` attribute does. 
>
> Compared to the simplicity of understanding how different **covariates** contribute towards the final **predicted values** of **multiple linear regression models** (by just reading off the equation to see how predictions work), the the complexity of how all the different **features** interact and combine to together to create the final **predictions** from **classification decision trees** can be staggering. But the so-called **feature importance** heuristics allows us to judge how relatively important the overall contributions from different features are in the final decision tree predictions. Now we just need to be sure we're not **overfitting** our **classification decision trees** since they can be so **complex**. Fortunately, the "GridSearchCV" methodology mentioned in regards to finding the best `max_depth` setting for a tree is going to provide a general answer to the challenge of complexity and **overfitting** in **machine learning models** that is not too hard to understand (and which you might already have some guesses or a hunch about).

````python
import numpy as np

# Retrieve feature importances from clf2
feature_importances = clf2.feature_importances_
features = clf2.feature_names_in_  # Corresponding feature names

# Visualize feature importances
plt.figure(figsize=(8, 6))
plt.barh(features, feature_importances, color='skyblue')
plt.xlabel("Importance Score")
plt.ylabel("Feature")
plt.title("Feature Importances for clf2")
plt.show()

# Report the most important feature
most_important_feature = features[np.argmax(feature_importances)]
print(f"The most important predictor variable for clf2 is: {most_important_feature}")
````
